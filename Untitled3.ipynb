```python
from github import Github
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter
```

```python
g = Github()
repo = g.get_repo(input("Укажите путь до репозитория в формате Name/Repo: "))
```

```python
print(f"Название: {repo.name}")
print(f"Описание: {repo.description}")
print(f"Звёзды: {repo.stargazers_count}")
print(f"Forks: {repo.forks_count}")
```

```python
contents = repo.get_contents("")
for content in contents:
    print(f"{content.type}: {content.name}")
```

```python
commits = repo.get_commits()
for commit in commits[:5]:
    print(f"{commit.sha[:8]}: {commit.commit.message}")
```

```python
tags = list(repo.get_tags())
```

```python
branches = repo.get_branches()
for branch in branches:
    print(f"Ветка: {branch.name}")
```

```python
commits = repo.get_commits()
authors = []
dates = []

for commit in commits[:100]:
    if commit.author:
        authors.append(commit.author.login)
        dates.append(commit.commit.author.date)

date_counts = Counter([d.date() for d in dates])
dates_sorted = sorted(date_counts.items())

x_labels = [d.strftime('%d.%m') for d, _ in dates_sorted]
x_positions = range(len(x_labels))
y_values = [count for _, count in dates_sorted]

plt.figure(figsize=(12, 5))
plt.bar(x_positions, y_values, width=0.7, alpha=0.7)

plt.xticks(x_positions, x_labels, rotation=45)

plt.title("Коммиты по дням")
plt.xlabel("Дата")
plt.ylabel("Количество коммитов")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
def get_all_files(repo, path=''):
  all_files = []
  try:
      contents = repo.get_contents(path)
      for content in contents:
          if content.type == "dir":
              all_files.extend(get_all_files(repo, content.path))
          else:
              all_files.append(content)
  except Exception as e:
      print(f"Ошибка при получении содержимого {path}: {e}")
  return all_files
def categorize_extension(ext):
    ext = ext.lower()

    programming_langs = {
        'py': 'Python', 'js': 'JavaScript', 'ts': 'TypeScript',
        'java': 'Java', 'cpp': 'C++', 'c': 'C', 'cs': 'C#',
        'go': 'Go', 'rs': 'Rust', 'php': 'PHP', 'rb': 'Ruby',
        'swift': 'Swift', 'kt': 'Kotlin', 'scala': 'Scala'}

    web = {'html', 'htm', 'css', 'scss', 'sass', 'less'}
    config = {'json', 'yml', 'yaml', 'toml', 'ini', 'xml', 'md', 'txt'}
    data = {'csv', 'tsv', 'jsonl', 'parquet'}
    docs = {'md', 'rst', 'txt', 'pdf'}
    images = {'png', 'jpg', 'jpeg', 'gif', 'svg', 'ico'}
    notebooks = {'ipynb'}

    if ext in programming_langs:
        return 'Программирование'
    elif ext in web:
        return 'Веб-файлы'
    elif ext in config:
        return 'Конфигурация/Документация'
    elif ext in data:
        return 'Данные'
    elif ext in notebooks:
        return 'Ноутбуки'
    elif ext in docs:
        return 'Документация'
    elif ext in images:
        return 'Изображения'
    elif ext in {'gitignore', 'dockerfile', 'license'}:
        return 'Служебные файлы'
    else:
        return 'Другие'

all_files = get_all_files(repo)
categories = []

for file in all_files:
    if '.' in file.name:
        ext = file.name.split('.')[-1].lower()
        categories.append(categorize_extension(ext))
    else:
        categories.append('Без расширения')

category_counts = Counter(categories)

plt.figure(figsize=(10, 10))
plt.pie(category_counts.values(),
        labels=category_counts.keys(),
        autopct=lambda pct: f'{pct:.1f}%\n({int(pct*len(all_files)/100)})',
        startangle=90)
plt.title(f'Распределение файлов по категориям\nВсего файлов: {len(all_files)}')
plt.tight_layout()
plt.show()
```

```python
tags = list(repo.get_tags())

if len(tags) == 0:
    print("⚠️ Тегов не найдено")
else:
    data = []
    for tag in tags:
        try:
            if tag.commit and tag.commit.commit and tag.commit.commit.author:
                data.append({
                    'Релиз': tag.name,
                    'Дата': tag.commit.commit.author.date,
                    'Автор': tag.commit.commit.author.name
                })
            else:
                data.append({
                    'Релиз': tag.name,
                    'Дата': None,
                    'Автор': 'Неизвестен'
                })
        except Exception as e:
            print(f"Ошибка с тегом {tag.name}: {e}")
            data.append({
                'Релиз': tag.name,
                'Дата': None,
                'Автор': 'Ошибка'
            })

    df = pd.DataFrame(data)

    df_with_dates = df.dropna(subset=['Дата'])

    if len(df_with_dates) == 0:
        print("❌ Нет тегов с информацией о дате")
    else:
        df_with_dates = df_with_dates.sort_values('Дата')

        print("📋 Последние 10 релизов:")
        print(df_with_dates.tail(10)[['Дата', 'Релиз', 'Автор']].to_string(index=False))

        plt.figure(figsize=(10, 4))
        plt.plot(df_with_dates['Дата'], range(len(df_with_dates)), 'o-', alpha=0.7)
        plt.title(f'История релизов {repo.name}')
        plt.xlabel('Дата')
        plt.ylabel('Порядковый номер релиза')
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()
```
